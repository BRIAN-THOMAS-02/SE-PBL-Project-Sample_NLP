### Data Pre-Processing

```python
import string
text = open('Testing.txt', encoding= 'utf-8').read()
lower_case = text.lower()
refined_text = lower_case.translate(str.maketrans('','',string.punctuation))

print("Text in Lower Case : \n\n" + lower_case, '\n\n\n')
print("Text without Punctuation : ", refined_text)
```

### Tokenization

```python
tokenized_words = refined_text.split()
print(tokenized_words)
```

### Stop Words - These are the words to avoid or remove from the Raw Text (Easy for Tokenization)

```python
global stop_words
stop_words = ["i", "im", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself",
              "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself",
              "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these",
              "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do",
              "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while",
              "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before",
              "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again",
              "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each",
              "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than",
              "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]
```

```python
print(stop_words)
```

### Removing Stop Words from Raw-Text

```python
pre_processed_text = []

for word in tokenized_words:
    if word not in stop_words:
        pre_processed_text.append(word)
        
print(pre_processed_text)
```

#### Refining the Output Visually:

#### Appending all the Words with corresponding Emotions, so that the emotions.txt file need not be opened every time

```python
words_list = []
emotions_list = []
with open('emotions.txt', 'r') as file:
    for line in file:
        refined_line = line.replace('\n','').replace(',','').replace("'",'')
        words, emotions = refined_line.split(':')
        words_list.append(words)
        emotions_list.append(emotions)
        
        
print(" Words       | Corresponding Emotions ")        
for i,j in zip(words_list,emotions_list):
    print(i, " | ", j)
```
